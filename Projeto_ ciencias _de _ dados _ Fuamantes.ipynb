```python
# Importando as bibliotecas
import polars as pl
```

```python

# URL do arquivo CSV bruto no GitHub
url = "https://raw.githubusercontent.com/ProfMarceloRodrigues/Projeto-de-Ciencias-de-Dados-de-Fumantes/main/tip.csv"

# Ler a base de dados diretamente da URL
tabela = pl.read_csv(url)

# Exibir a tabela
print(tabela)

```

```python
tabela = tabela.drop_nulls()
```

```python
display(tabela.schema)
```

```python
display(tabela.null_count()) 
```

```python
import seaborn  as  sns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="sex", y="size", data=tabela, palette="pastel")  # muda a paleta de cores
plt.title("Distribuição do tamanho por sexo")                  # adiciona título
plt.xlabel("Sexo")                                              # opcional: renomeia eixo X
plt.ylabel("Tamanho")                                           # opcional: renomeia eixo Y
plt.show()


```

```python
# A distribuição de tamanho entre fumantes é semelhante para homens e mulheres, com medianas próximas.
#Ambos os sexos apresentam variação moderada e alguns outliers acima da média.
#-Isso sugere que o tabagismo pode influenciar o tamanho de forma parecida entre os sexos.
```

```python
sns.set_style("whitegrid")

sns.scatterplot(x="time", y="total_bill", hue="sex", data=tabela, palette="Set2")

plt.title("Valor da conta por período do dia", fontsize=14, fontweight="bold")
plt.xlabel("Período (Almoço ou Jantar)", fontsize=12)
plt.ylabel("Valor total da conta (R$)", fontsize=12)

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

```python
#Fumantes são mais frequentes no jantar, período com as contas mais altas.
#Homens fumantes dominam os maiores valores de consumo, indicando refeições mais caras ou grupos maiores.
#Mulheres fumantes mantêm padrão mais moderado, com menor presença em contas elevadas.
```
